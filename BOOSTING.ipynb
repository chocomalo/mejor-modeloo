```python
!pip install wooldridge
import wooldridge as wd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
import wooldridge as wd
hprice = wd.data('hprice1')
```

```python
hprice.info()
```

```python
hprice = hprice [['price', 'assess', 'bdrms', 'lotsize', 'sqrft', 'colonial']]
```

```python
y = hprice ['price']
x = hprice [['assess', 'bdrms', 'lotsize', 'sqrft', 'colonial']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)
```

#CatBoost ( MEJOR MODELO)

```python
param_grid = {'n_estimators':[85],
              'max_depth': [2],}
rf_grid1 = GridSearchCV (CatBoostRegressor(), param_grid=param_grid, cv=8, scoring='neg_mean_squared_error')
rf_grid1.fit (x_train, y_train)
```

```python
rf_grid1.best_score_
```

#GradientBoosting

```python
param_grid = {'n_estimators':[85],
              'max_depth': [2],}
rf_grid2 = GridSearchCV (GradientBoostingRegressor(), param_grid=param_grid, cv=8, scoring='neg_mean_squared_error')
rf_grid2.fit (x_train, y_train)
```

```python
rf_grid2.best_score_
```

#HistGradientBoosting

```python
param_grid = {'max_depth': [2],}
rf_grid3 = GridSearchCV (HistGradientBoostingRegressor(), param_grid=param_grid, cv=8, scoring='neg_mean_squared_error')
rf_grid3.fit (x_train, y_train)
```

```python
rf_grid3.best_score_
```

#XGBoosting

```python
from xgboost.sklearn import XGBRegressor
param_grid = {'n_estimators':[85],
              'max_depth': [2],}
rf_grid4 = GridSearchCV (XGBRegressor(), param_grid=param_grid, cv=8, scoring='neg_mean_squared_error')
rf_grid4.fit (x_train, y_train)
```

```python
rf_grid4.best_score_
```

#LightGBM

```python
param_grid = {'n_estimators':[85],
              'max_depth': [2],}
rf_grid5 = GridSearchCV (LGBMRegressor(), param_grid=param_grid, cv=8, scoring='neg_mean_squared_error')
rf_grid5.fit (x_train, y_train)
```

```python
rf_grid5.best_score_
```
